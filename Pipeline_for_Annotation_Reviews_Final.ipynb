```python
# @title Visualise
import umap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Function to generate Cohere embeddings
def embed_text(texts):
    embeddings = []
    for i in range(0, len(texts), 90):
        batch = texts[i:i+90]
        output = co.embed(
            model="embed-english-v3.0",
            texts=batch,
            input_type="clustering")
        embeddings.extend(output.embeddings)
    return embeddings

# Function to reduce dimensionality of embeddings using UMAP directly to 2 dimensions
def reduce_dimensionality(embeddings, n_components=2):
    reducer = umap.UMAP(n_components=n_components, n_neighbors=3, init='random', min_dist=0.1)
    umap_embeddings = reducer.fit_transform(embeddings)
    return umap_embeddings

# Normalise the data
def normalize_data(data):
    scaler = MinMaxScaler()
    normalized_data = scaler.fit_transform(data)
    return normalized_data

# Get texts from the dataframe
texts = final_feedback_df['Text'].tolist()

# Generate embeddings
embeddings = embed_text(texts)

# Reduce dimensionality to 2 dimensions
umap_embeddings_2d = reduce_dimensionality(embeddings, n_components=2)

# Normalise embeddings
normalized_embeddings_2d = normalize_data(umap_embeddings_2d)

# Convert embeddings to numpy array for easier manipulation
umap_embeddings_np_2d = np.array(normalized_embeddings_2d)

# Add reduced dimensions to the dataframe
final_feedback_df['dim1'] = umap_embeddings_np_2d[:, 0]
final_feedback_df['dim2'] = umap_embeddings_np_2d[:, 1]

# Plot the reduced dimensions with 'Category' as a legend
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='dim1',
    y='dim2',
    hue='Category',
    palette='viridis',
    data=final_feedback_df,
    legend='full'
)
plt.title('2D UMAP Projection of Text Embeddings')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend(title='Category')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the elbow method
def plot_elbow_method(data, max_clusters=10):
    wcss = []  # Within-cluster sum of squares
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, random_state=42, n_init='auto')
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_clusters + 1), wcss, marker='o')
    plt.title('Elbow Method For Optimal Number of Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()

# Generate KMeans clusters
def generate_clusters(data, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    cluster_labels = kmeans.fit_predict(data)
    return cluster_labels

# Plot the elbow chart
plot_elbow_method(normalized_embeddings_2d)

# Choose the optimal number of clusters (e.g., 3 from the elbow method)
optimal_clusters = 7

# Generate clusters
final_feedback_df['Cluster'] = generate_clusters(normalized_embeddings_2d, optimal_clusters)

# Calculate silhouette score
sil_score = silhouette_score(normalized_embeddings_2d, final_feedback_df['Cluster'])
print(f'Silhouette Score: {sil_score}')

# Visualize the clusters
fig = px.scatter(
    final_feedback_df,
    x='dim1',
    y='dim2',
    color='Category',
    hover_data=['Text', 'Category'],
    title='2D UMAP Projection of Text Embeddings with KMeans Clusters'
)

fig.update_layout(
    xaxis_title='Dimension 1',
    yaxis_title='Dimension 2'
)

fig.show()
```

```python
# Preamble for the Cohere model
preamble = """
You are an expert in categorizing and summarizing text data. Given a list of text items, your task is to generate a label that best describes the common theme within the cluster of texts.
Each cluster represents feedback items that share a similar theme or topic. Provide a concise and clear label that accurately captures the essence of the cluster.
"""
```

```python
# Function to get cluster labels using Cohere's model
def get_cluster_label(cluster_texts, preamble):
    # Create a bullet list string
    bullet_list = "\n".join([f"- {text}" for text in cluster_texts])

    # Prepare the message for the model
    message = f"Here is a list of feedback items:\n{bullet_list}\n\nPlease provide a label that best describes the items in the list."

    response = co.chat(
        model='command-r-plus',
        message=message,
        temperature=0.2,
        preamble=preamble,
        prompt_truncation='AUTO'
    )
    return response.text.strip()

# Group texts by cluster and generate labels
cluster_labels = {}
for cluster in filtered_feedback_df['Cluster'].unique():
    cluster_texts = filtered_feedback_df[filtered_feedback_df['Cluster'] == cluster]['Text'].tolist()
    print(f"Cluster {cluster} Texts: {cluster_texts}\n")
    label = get_cluster_label(cluster_texts, preamble)
    cluster_labels[cluster] = label
    print(f"Cluster {cluster} Label: {label}\n")

# Append the new label to the dataframe
filtered_feedback_df['Cluster_Label'] = filtered_feedback_df['Cluster'].map(cluster_labels)

# Display the dataframe with the new labels
filtered_feedback_df.head()
```
